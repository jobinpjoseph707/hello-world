```python
import os
import numpy as np
from matplotlib import pyplot as plt
import pywt
import cv2 
import pandas as pd

data=[]
basepath='./dataset/'
for folder in os.scandir(basepath):
    if folder.is_dir():
        #print(folder.name)
        
        for file in os.scandir(folder.path):
            features=[]
            im=cv2.imread(file.path,cv2.IMREAD_COLOR)
            im = cv2.resize(im,(150,150))
            plt.subplot(121), plt.imshow(im)
            im_gray = cv2.cvtColor( im,cv2.COLOR_BGR2GRAY)
            ret,t_im = cv2.threshold(im_gray,110,255,cv2.THRESH_BINARY)
            kernel = np.ones((1,1),np.uint8)
            t_im = cv2.morphologyEx(t_im, cv2.MORPH_OPEN, kernel)
            contours1,hierarchy = cv2.findContours(t_im, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
            cnt1=contours1[1]
            perimeter= cv2.arcLength(cnt1,True)
            features.append(perimeter)
            features.append(len(contours1))
            features.extend(np.mean(im, axis=(0, 1)).tolist())
            features.append(folder.name)
            data.append(features)     
print(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
df = pd.DataFrame(data, columns=['f','c','R','G','B','class'])
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(X)
print(y)

le = preprocessing.LabelEncoder()
y=le.fit(y).transform(y)

print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
